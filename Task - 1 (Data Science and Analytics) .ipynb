# Basic Visualizations

```python
# scatter plot to analyze relationships between variables.
sn.scatterplot(data=iris, x='petal_length', y='petal_width',
            hue='species', palette=['deeppink', 'blue', 'lime'])
```

Scatter plot shows that as petal length increases, petal width also increases — meaning longer petals tend to be wider too. Two separate groups:
a small cluster on the bottom-left (small petals) and a larger spread on the right ( bigger petals).

```python
#histogram to visulize the distribution

sn.histplot(data=iris, x="petal_length", hue="species",
             palette=['red', 'black', 'blue'], bins=20, kde=True)
plt.title("Histogram Petal Length Distribution")
plt.xlabel("Petal Length")
plt.ylabel("Count")

```

The histogram shows that Setosa (red) has very small petals (around 1-2 cm) and is completely separate from the other two. Versicolor (black) and Virginica (blue) have bigger petals but their ranges overlap a little with each other.

```python
#boxplot to compare distributions of petal length across differnt spscies.

sn.boxplot(data=iris, x="species", y="petal_length",
            hue="species", palette=["deeppink", "blue", "lime"])
plt.title("Box Plot - Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length")
plt.show()
```

The box plot shows that Setosa (green) has the smallest petals (around 1-2 cm) with very little variation, while Virginica (blue) has the longest petals. The small circles you see on Setosa and Versicolor are outliers, meaning a few flowers had unusual petal lengths compared to the rest of their species.
